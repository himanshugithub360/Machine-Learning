### In Data Analysis we will analyze to Find out the below stuff

1. Missing Values
2. All the Numerical Variables
3. Distribution of the Numerical Variables
4. Categorical Variables
5. Cardinality of categorical Variables
6. Outliers
7. Relationship between independent and dependent feature(Sale Price)

```python
# Here we will check the percentage of nan values present in each feature
# I - Step make the list of features which has missing values

features_with_na = [features for features in Dataset.columns if Dataset[features].isnull().sum()>1]

# II - step print the feature name and the percentage of missing values

for features in features_with_na:
    print(features,np.round(Dataset[features].isnull().mean(), 4),"% missing values")
```

### Since they are many missing values, we need to find the relationship between missing values and Sales Price

let's plot some diagram for this relationship

```python
data = Dataset.copy()
for features in features_with_na:
    # lets make a variable that indicates 1 if the obsevation was missing or else 0
    data[features] = np.where(data[features].isnull(), 1,0)

# lets calculate the mean SalePrice where the information is missing or present
   # Group data
    grouped_data = data.groupby(features)["SalePrice"].median()

    # Assign colors based on index (0  orange, 1  blue)
    colors = ['orange' if i == 0 else 'blue' for i in grouped_data.index]

    # Plot
    grouped_data.plot.bar(color=colors)
    plt.title(features)
    plt.show()    
```

Here with the relation between the missing values and the dependent variables is clearly visible. so We need to replace these nan values with something meaningful which we will do in the Feature Engineering section.

### Numerical Variables

```python
#list of numerical variables

numerical_features = [feature for feature in Dataset.columns if Dataset[feature].dtype != 'O']

print("Number of numerical variables : ", len(numerical_features))

#Visual the numerical variables

Dataset[numerical_features].head()
```

### Temporal Variables(Eg: Datetime Variables)

 

```python
# list of variables that contain yeat information

year_feature = [feature for feature in numerical_features if "Yr" in feature or "Year" in feature]

year_feature
```

```python
# lets explore the content of these year variables
for feature in year_feature:
    print(feature, Dataset[feature].unique()) 
```

```python
# lets analyze the Temporal datatime Variables
# we will check whether there is a relation between year the house is sold and the Saleprice

Dataset.groupby("YrSold")["SalePrice"].median().plot()
plt.xlabel("Year Sold")
plt.ylabel("Median House Price")
plt.title("House Price VS YearSold")
```

```python
# Here we will compare the differnece between all year feature with SalePrice

for feature in year_feature:
    if feature != "YrSold":
        data = Dataset.copy()
        data[feature] = data["YrSold"] - data[feature]

        plt.scatter(data[feature], data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("Saleprice") 
        plt.show()
```

```python
# Numerical variables are usally of 2 types
# 1. Continous variables  2. Discrete variables

discrete_feature = [feature for feature in numerical_features if len(Dataset[feature].unique())<25 and feature not in year_feature+["Id"]]

print("Discrete Vaiables count : {}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
Dataset[discrete_feature].head() 
```

```python
# lets Find the relationship between them and salePrice

for feature in discrete_feature:
    data= Dataset.copy()
    data.groupby(feature)["SalePrice"].median().plot.bar(color= "b")
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title(feature)
    plt.show()
```

### Continous Variables

```python
continous_feature = [feature for feature in numerical_features if feature not in discrete_feature + year_feature + ["Id"]]

print("Continous feature Count : {}".format(len(continous_feature)))
```

```python
continous_feature
```

```python
# Lets analyse the continuous values by creating histogram to understand the count and the feature


for feature in continous_feature:
    data = Dataset.copy()
    data[feature].hist(bins = 25)
    plt.xlabel(feature)
    plt.ylabel("count")
    plt.title(feature)
    plt.show()
```

```python
# we will be using logarithmic transformation

for feature in continous_feature:
    data = Dataset.copy()
    if 0 in Dataset[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data["SalePrice"] = np.log(data["SalePrice"])
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("SalesPrice")
        plt.title(feature)
        plt.show()
```

### Outliers

```python
for feature in continous_feature:
    data = Dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
                  
```
