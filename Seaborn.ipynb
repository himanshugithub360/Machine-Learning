# Seaborn
---

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df
```

```python
df.dtypes
```

### Correlation with Heatmap
---
A correlation with heatmap uses colored cells, typically in a monochromatic scale,to show a 2D a correlation matrix (table) between two discrete dimension or the events types.

```python
df.corr(numeric_only= True)
```

```python
sns.heatmap(df.corr(numeric_only= True))
```

### Jointplot
---
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

#### Univariate Analysis

```python
sns.jointplot(x='tip', y='total_bill',data=df, kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
# using reg it also draw the regression line for it
```

#### Pairplot
A pairplot is also Know as a Scatterplot, in which one variable of same data row is matched with another varible's value

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

### Dist plot
 It helps to check the distribution of the columns features. used to plot the distribution of a dataset (histogram + KDE).

```python
sns.distplot(df["tip"])
```

```python
sns.displot(df["tip"],kde=False,bins=10)
```

## Categorial Plots
---
Seaborn also helps us in doing the analysis on Categorial Data points
- Boxplot
- Violinplot
- Countplot
- Barplot 

```python
# Count Plot
sns.countplot(x ='smoker',data= df, hue= 'sex')
```

```python
sns.countplot(x ='day',data= df)
```

```python
sns.countplot(y ='day',data= df)
```

```python
# Bar Plot
sns.barplot(x ='total_bill', y ='smoker',data= df)
```

```python
# Box Plot
sns.boxplot(x ='sex',y = 'total_bill',data = df)
```

```python
sns.boxplot(x ='sex',y = 'total_bill',data = df, palette="rainbow")
```

```python
#               Violin plot
#------------------------------------------------------------------------
# it helps us to see both the distribution of the data in term of Kernel density estimation and the box plot.
 
sns.violinplot(x="day",y='total_bill', data=df, palette = "rainbow")
```
