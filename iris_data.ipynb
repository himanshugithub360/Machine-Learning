## Univariate Analysis

```python
df_setosa=df.loc[df["species"] == "setosa"]
df_setosa
```

```python
df_virginica = df.loc[df["species"] == "virginica"]
df_versicolor = df.loc[df["species"] == "versicolor"]

```

```python
plt.plot(df_setosa['petalLength'], np.zeros_like(df_setosa['petalLength']),'o')
```

```python
plt.plot(df_versicolor['petalLength'], np.zeros_like(df_versicolor['petalLength']),'o',color= 'red')
```

```python
plt.plot(df_virginica['petalLength'], np.zeros_like(df_virginica['petalLength']),'o',color= "green")
```

```python
plt.plot(df_setosa['petalLength'], np.zeros_like(df_setosa['petalLength']),'o')
plt.plot(df_versicolor['petalLength'], np.zeros_like(df_versicolor['petalLength']),'o',color= 'red')
plt.plot(df_virginica['petalLength'], np.zeros_like(df_virginica['petalLength']),"o",color= "green")
```
