```python
from  sklearn.datasets import load_breast_cancer
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = load_breast_cancer()
```

```python
df
```

```python
Dataset = pd.DataFrame(df.data)
Dataset
```

```python
Dataset.columns = df.feature_names
```

```python
Dataset
```

```python
df.target.shape
```

```python
Dataset["price"] = df.target
```

```python
Dataset.head()
```

```python
Dataset["price"].value_counts()
```

```python
X = Dataset.iloc[ : , : -1]
Y = Dataset.iloc[ : , -1]
plt.plot(X,Y)
```

## Linear Regression

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

lin_regressor = LinearRegression()
mse = cross_val_score(lin_regressor,X,Y,scoring="neg_mean_squared_error",cv=8)
mean_mse = np.mean(mse)
print(mean_mse)
```

## Ridge Regression


```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

ridge = Ridge()
parameter = {'alpha' : [1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}

ridge_regressor = GridSearchCV(ridge, parameter, scoring= 'neg_mean_squared_error',cv=5)
ridge_regressor.fit(X,Y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

## Lasso Regression

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso

lasso = Lasso()
parameter = {'alpha' : [1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regressor = GridSearchCV(lasso, parameter,scoring = "neg_mean_squared_error",cv = 5)

lasso_regressor.fit(X,Y)

```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```
