## MultiCollinearity in Linear Regression

```python

import pandas as pd
import statsmodels.api as sm
```

```python
df_adv = pd.read_csv("Dataset/Advertisment.csv")
df_adv

```

```python
X = df_adv[["TV","Radio","Newspaper"]]
Y = df_adv["Sales"]
```

```python
X
```

```python
X = sm.add_constant(X)
X
```

```python
# fit a OLS Model with intercept on TV and Ratio
model = sm.OLS(Y,X).fit()
```

```python
model.summary()
```

```python
X.iloc[:, 1:].corr()
```

```python
df_salary = pd.read_csv("Dataset/Salary_Data.csv")
df_salary
```

```python
X = df_salary[["Age", "Years of Experience"]]
Y = df_salary['Salary']
```

```python
X = sm.add_constant(X)
X.isnull().sum()
```

```python
import numpy as np
import pandas as pd

print("NaN values in X:\n", pd.DataFrame(X).isna().sum())
print("NaN values in Y:\n", pd.Series(Y).isna().sum())

print("Any inf in X:", np.isinf(X).any())
print("Any inf in Y:", np.isinf(Y).any())
```

```python
X = X.dropna(inplace=True)
Y = Y.dropna(inplace=True)
```

```python
mask = ~np.isnan(X).any(axis=1) & ~np.isnan(Y)
X = X[mask]
Y = Y[mask]
```

```python
  # add constant for intercept
model = sm.OLS(Y, X).fit()
print(model.summary())
```

```python
X.iloc[:,1:].corr()
```
