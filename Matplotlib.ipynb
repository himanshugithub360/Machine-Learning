```python
import matplotlib.pyplot as plt

%matplotlib inline

# second line is written for plt.show() will be automatically called , doesnot need to specific call
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel("Y axis")
plt.title("Graph in 2D")
plt.savefig("Graph.png")
```

```python
y = x*x
```

```python
plt.plot(x,y,"r*",linestyle= "dashed",linewidth=2, markersize=15)
```

```python
# Creating Subploting
plt.subplot(2,2,1)
plt.plot(x,y,"ro")
plt.subplot(2,2,2)
plt.plot(x,y,"b--")
plt.subplot(2,2,3)
plt.plot(x,y,"g")
```

```python
np.pi
```

```python
# Computing the X and Y coordinates for points on a sine curve
x = np.arange(0, 4*np.pi, 0.2 )
y = np.sin(x)
plt.plot(x,y)
plt.title("sine wave form") 
```

```python
x = np.arange(0,5*np.pi, 0.2)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.title("sine wave")
plt.plot(x,y_sin,"g--")

plt.subplot(2,1,2)
plt.title("cose wave")
plt.plot(x,y_cos,"r--")
```

```python
# Bar Plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2) 
plt.legend()
```

```python
# Histogram

a = np.array([12,32,43,45,56,58,87,54,2,12,1,23,32,64,76,56,23,78,89])
plt.hist(a)
```

## Box plot using Matplot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)

# vert represent vertical representation 
# patch_artist represent color of box
```

## Pie chart

```python
labels = 'Python', 'C','C++','Java'
sizes = 213, 130, 260,300 
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0) 

plt.pie(sizes,explode,labels,colors , autopct ='%1.2f%%',shadow=True )
plt.axis('equal')
```
